```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
import random

def get_comments():
    def extract_comments():
        page_comments = []
        comments = driver.find_elements(By.CSS_SELECTOR, 'div.shopee-product-rating')
        for comment in comments:
            # AUTHOR OF THE COMMENT
            comment_author = comment.find_element(By.CLASS_NAME, 'shopee-product-rating__author-name').text.strip()
            
            # TIME AND CATEGORY
            time_and_category = comment.find_element(By.CLASS_NAME, 'shopee-product-rating__time').text.strip()
            
            
            # CATEGORIZED COMMENTS
            time_element = comment.find_element(By.CLASS_NAME, 'shopee-product-rating__time')
            next_div_element = time_element.find_element(By.XPATH, 'following-sibling::div')

            categorized_feedbacks_dict = {}
            for i in next_div_element.find_elements(By.CSS_SELECTOR, 'div:not([style*="margin-top: 0.75rem"])'):
                if i.text == 'Phản Hồi Của Người Bán':
                    break
                if len(i.text.split(': ')) == 2: 
                    category, feedback = i.text.split(': ')[0], i.text.split(': ')[1]
                    categorized_feedbacks_dict[category] = feedback
                
            
            # FULL COMMENT = Bình luận có phân loại + bình luận viết tay
            time_element = comment.find_element(By.CLASS_NAME, 'shopee-product-rating__time')
            next_div_element = time_element.find_element(By.XPATH, 'following-sibling::div')
            full_comment = next_div_element.text.strip()
            if 'Phản Hồi Của Người Bán' in full_comment:
                full_comment = ''
            
            # RATING
            svg_elements = comment.find_elements(By.CSS_SELECTOR, '.shopee-product-rating__rating svg')
            count_fill_none = sum(1 for svg in svg_elements if svg.find_element(By.CSS_SELECTOR, 'polygon').get_attribute('fill') == 'none')
            stars = 5 - count_fill_none
            
            
            page_comments.append({'Author': comment_author, 'Time and Category': time_and_category, 'Categorized Feedbacks': categorized_feedbacks_dict, 'Stars': stars, 'Full comment': full_comment})
        return page_comments
    
    def end_page():
        buttons_container = driver.find_element(By.CSS_SELECTOR, 'div.shopee-page-controller.product-ratings__page-controller')
        
        buttons = buttons_container.find_elements(By.TAG_NAME, 'button')
        
        next_button_selector = "#main > div > div:nth-child(3) > div:nth-child(1) > div > div > div > div.wAMdpk > div > div.page-product__content--left > div:nth-child(2) > div > div > div.product-ratings__list > div.shopee-page-controller.product-ratings__page-controller > button.shopee-icon-button.shopee-icon-button--right"
        next_button = driver.find_element(By.CSS_SELECTOR, next_button_selector)
        
        next_button_index = buttons.index(next_button)
        
        if next_button_index > 0 and "shopee-button-solid--primary" in buttons[next_button_index - 1].get_attribute('class'):
            return True
        else:
            return False
        
        
    

    ###### MAIN #####
    all_comments_list = []
    product_info_section = driver.find_element(By.XPATH, "//h2[text()='Product Information Section']/following-sibling::div")
    product_info_section.find_element(By.CSS_SELECTOR, 'button.flex:nth-child(2)').click()

    counter = 1
    while True:
        delay = random.uniform(5, 7)
        time.sleep(delay)
        page_comments = extract_comments()
        all_comments_list.extend(page_comments)  
            
        if end_page() == True:
            print(f'\rFinished {counter} page.', end='')
            print('\nLast page.')
            break
        else:
            print(f'\rFinished {counter} page.', end='')
            next_page_button_css_selector = '#main > div > div:nth-child(3) > div:nth-child(1) > div > div > div > div.wAMdpk > div > div.page-product__content--left > div:nth-child(2) > div > div > div.product-ratings__list > div.shopee-page-controller.product-ratings__page-controller > button.shopee-icon-button.shopee-icon-button--right'
            next_page = driver.find_element(By.CSS_SELECTOR, next_page_button_css_selector)
            next_page.click()
        counter += 1
    return all_comments_list
```
